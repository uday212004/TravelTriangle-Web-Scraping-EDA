```python
df['Package Original Price ₹'] = (
    df['Package Original Price ₹']
    .str.replace('₹','')
    .str.replace(',','')
    .str.replace('/-','')
    .astype(float)
)
```

```python
df['Hotel Rating'] = df['Hotel Rating'].str.extract('(\d+)').astype(float)
```

```python
df['Hotel Rating'].fillna(df['Hotel Rating'].median(), inplace=True)
```

```python
df
```

```python
df['Days'] = df['Package Duration'].str.extract('(\d+)').astype(float)
```

```python
df
```

```python
df['Nights'] = (
    df['Package Duration']
    .str.extract(r'(\d+)\s*Nights')
    .astype(float)
)
```

```python
df
```

```python
df.columns
```

```python
df['Discount %'] = df['Package Discount %'].str.replace('% Off','').astype(float)
```

```python
df
```

```python
df
```

```python
df.describe()
```

```python
df['Package'] = df['Package'].str.strip()
```

```python
df
```

```python
df['Discount %'] = df['Package Discount %'].str.replace('% Off','').astype(float)
```

```python
df
```

```python
df.drop(columns=['Package Discount %'], inplace=True)
```

```python
df
```

```python
df['City'] = (
    df['City']
    .str.replace('Cities:', '', regex=False)
    .str.replace(r'\(\d+D\)', '', regex=True)
    .str.strip()
)
```

```python
df
```

```python
#df['Primary City'] = df['City'].str.split(',').str[0]

df.describe()
```

```python
df
```

```python
df['Activity'] = df['Activity'].str.replace(
    r'(?<!^)(?=[A-Z])', ' ', regex=True
)
```

```python
df
```

```python
df['Activity Count'] = df['Activity'].str.split().apply(len)
```

```python
df
```

```python
df.rename(
    columns={'Package Original Price ₹': 'Package Price'},
    inplace=True
)
```

```python
df
```

```python
df['Price per Day'] = df['Package Price'] / df['Days']
```

```python
df
```

```python
df['Final Price'] = df['Package Price'] * (1 - df['Discount %']/100)
```

```python
plt.figure()
plt.hist(df['Package Price'], bins=10)
plt.xlabel("Package Price (₹)")
plt.ylabel("Number of Packages")
plt.title("Distribution of Travel Package Prices")
plt.show()
```

```python
df['City'].value_counts().head(12).plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Number of Packages")
plt.title("Top 10 Cities by Number of Packages")
plt.show()
```

```python
df['Hotel Rating'].value_counts().sort_index().plot(kind='bar')
plt.xlabel("Hotel Rating")
plt.ylabel("Number of Packages")
plt.title("Hotel Rating Distribution")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Days'], df['Package Price'])
plt.xlabel("Number of Days")
plt.ylabel("Package Price (₹)")
plt.title("Price vs Trip Duration")
plt.show()
```

```python
df.groupby('Hotel Rating')['Package Price'].mean().plot(kind='bar')
plt.xlabel("Hotel Rating")
plt.ylabel("Average Price (₹)")
plt.title("Average Price by Hotel Rating")
plt.show()
```

```python
plt.hist(df['Discount %'], bins=20)
plt.xlabel("Discount %")
plt.ylabel("Count")
plt.title("Distribution of Discounts")
plt.show()
```

```python
plt.scatter(df['Discount %'], df['Package Price'])
plt.xlabel("Discount %")
plt.ylabel("Package Price (₹)")
plt.title("Discount vs Package Price")
plt.show()
```

```python
plt.boxplot(df['Package Price'])
plt.ylabel("Package Price (₹)")
plt.title("Outlier Detection in Package Prices")
plt.show()
```

```python
plt.hist(df['Price per Day'], bins=30)
plt.xlabel("Price per Day (₹)")
plt.ylabel("Count")
plt.title("Distribution of Price per Day")
plt.show()
```
